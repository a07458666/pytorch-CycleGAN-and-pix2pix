```python
import open3d as o3d
import matplotlib.pyplot as plt
%matplotlib widget

depthPath = './datasets/depthEndoscopy/0003_prone/0003_prone_10506_depth.png'
colorPath = './datasets/depthEndoscopy/0003_prone/0003_prone_10506_image.png'

toCm = 16 / 65535 
```

```python
depth = plt.imread(depthPath)
color = plt.imread(colorPath)
```

```python
plt.subplot(1, 2, 1)
plt.title('color image')
plt.imshow(color)
plt.subplot(1, 2, 2)
plt.title('depth image')
plt.imshow(depth)
plt.show()
```

```python
import cv2
```

```python
depthImg = cv2.imread(depthPath, -1)
```

```python
print("max(cm) : {:.2f}cm".format(depthImg.max() * toCm), "min(cm) : {:.2f}cm".format(depthImg.min() * toCm))
```
